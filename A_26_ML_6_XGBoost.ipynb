```python
import numpy as np
import pandas as pd
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header = None)

test_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.test', skiprows = 1, header = None)

# header = None will make the top row skip which in this case is just col details.

col_labels = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status',
'occupation','relationship', 'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week',
'native_country', 'wage_class']

train_set.columns = col_labels

test_set.columns = col_labels
```

```python
# i noticed some "?" in our data. But info shows me non-null values.
# i will do dropna and see the result.

train_set.replace(' ?',np.nan).dropna().shape

# if i do '?', then same return, 32561 rows
# if i give a space like this ' ?', then 30162 rows returned. weird...      ' ?' vs '?'
```

```python
train_clean=train_set.replace(' ?',np.nan).dropna()
test_clean=test_set.replace(' ?',np.nan).dropna()

# now missing value has been taken care of.
```

```python
 # <50k.   , < 50 k    ????
```

```python
test_clean['wage_class'] = test_clean.wage_class.replace({' <=50K.': ' <=50K', ' >50K.':' >50K'})
```

```python
# for XGB we need to convert all strings to numeric form.
# all feature vectors shud be in digits only.
# we do this by ordinal encoding or one-hot-encoding.
# one-hot-encoding isn't necessary for tree-based algos.
# ordinal encoding uses less memory if we have multiple categorical features.
```

```python
# lets combine train and test set and do ordinal encoding on the required featres together.
train_test_combined=pd.concat([train_clean,test_clean],axis=0)
train_test_combined.head()
```

```python
# we can see above that both dataframes has been stacked one on top of other i e vertically
```

```python
train_test_combined.info()
```

```python
# now we do nominal encoding through for loop to all columns which are strings and categorical(non-numeric).
# they come up as 'pbject' in info instead of int64( see above)

for features in train_test_combined.columns:
    if train_test_combined[features].dtype=='object':
        train_test_combined[features] = pd.Categorical(train_test_combined[features]).codes  # this changes strings to integers.
```

```python
train_test_combined.info() # all object changed to int
```

```python
train_test_combined['race'][:20] # just to check if white,black,hispanic,asian-islanders etc categories has been converted to numerics.
```

```python
X_train=train_test_combined[:train_clean.shape[0]] # end row number of initial clean training set.
X_test=train_test_combined[train_clean.shape[0]:] # past the last row of the clean training set.
print(X_train.shape)
```

```python
# Data preprocessing and cleaning and conversions (into numeric from string) is done.
```

```python
y_train=X_train.pop('wage_class')
y_test=X_test.pop('wage_class')
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
xgbdmat=xgb.DMatrix(X_train,y_train) # create DMatrix to make XGB more efficient.
```

```python
params= {
    'objective':'binary:logistic',
    'max_depth':3,
    'silent':1,
    'eta':.2,
    'seed':18,
    'n_estimators':10
}
num_rounds= 525
bst=xgb.train(params,xgbdmat,num_rounds)
```

```python
testdmat = xgb.DMatrix(X_test)

from sklearn.metrics import accuracy_score
y_pred= bst.predict(testdmat)
y_pred
```

```python
# XGB outputs probabilities by default, not the actual class labels.
# we gotta convert to 0/1 label.

y_pred[y_pred > 0.5] = 1
y_pred[y_pred < 0.5] = 0
y_pred
```

```python
accuracy_score(y_pred, y_test)
```
